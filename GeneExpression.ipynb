```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import warnings
warnings.filterwarnings("ignore")
from sklearn.preprocessing import StandardScaler
```

```python
data1=pd.read_csv('/content/drive/MyDrive/Gene/data_set_ALL_AML_independent.csv')
data2=pd.read_csv('/content/drive/MyDrive/Gene/data_set_ALL_AML_train.csv')

#adding all the column headings to a list lst
lst=data1.columns.tolist()
lst_new=data2.columns.tolist()
dic= {}
for i in range(2,data1.shape[1],2):
  lst1=list(data1[lst[i]]) +list(data1[lst[i+1]])
  dic[int(lst[i])]=lst1
for i in range(2,data2.shape[1],2):
  lst1=list(data2[lst_new[i]]) +list(data2[lst_new[i+1]])
  dic[int(lst_new[i])]=lst1

k=list(dic.keys())
k.sort()
sorted_dic={str(i):dic[i] for i in k }
new_data=pd.DataFrame(sorted_dic)







```

```python
new_data=new_data.T
```

```python
new_data.replace('A',0,inplace=True)
new_data.replace('M',1,inplace=True)
new_data.replace('P',2,inplace=True)
```

```python
new_data
```

```python
data3=pd.read_csv('/content/drive/MyDrive/Gene/actual.csv')
data3
```

```python
mapping={'AML':0,'ALL':1}
data3['cancer']=data3['cancer'].map(mapping)
data3
```

```python
data3.drop('patient',axis=1,inplace=True)
```

```python
data3 = data3.squeeze()


data3.index = new_data.index


new_data['cancer'] = data3
```

```python
new_data
```

```python
x=new_data.drop('cancer',axis=1)
y=new_data['cancer']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import GridSearchCV


ridge = Ridge(alpha=1.0)
ridge.fit(x_train, y_train)



```

```python
lasso = Lasso(alpha=0.1)
lasso.fit(x_train, y_train)
```

```python

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

ridge_pred = (ridge.predict(x_test) >= 0.5).astype(int)


lasso_pred = (lasso.predict(x_test) >= 0.5).astype(int)
print("\nRidge Regression Results:")
print("Accuracy:", accuracy_score(y_test, ridge_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, ridge_pred))
print("Classification Report:\n", classification_report(y_test, ridge_pred))


print("\nLasso Regression Results:")
print("Accuracy:", accuracy_score(y_test, lasso_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, lasso_pred))
print("Classification Report:\n", classification_report(y_test, lasso_pred))
```
